```python
from sklearn import datasets
import pandas as pd

iris = datasets.load_iris()
print(iris)
```

```python
df = pd.DataFrame(iris["data"],columns = iris["feature_names"])
df["answer"] = iris["target"]
df.to_csv("iris.csv", encoding="utf-8", index = False)
df
```

```python
# step1 切資料, 會分成[ 90%題目(訓練), 10%題目(測試), 90%答案(訓練), 10%答案(測試) ]
# use train_test_split to split data, test_size is a ratial
from sklearn import model_selection
x_train, x_test, y_train, y_test = model_selection.train_test_split(iris["data"], iris["target"],test_size= 0.1)
print(x_train)
print(y_train)
```

```python
# step2 準備方法 
# new DecisionTreeClassifier
from sklearn import tree
clf = tree.DecisionTreeClassifier() 
```

```python
# step3 做考古題
# model.fit 
clf.fit(x_train, y_train)
```

```python
# step4 答題
# model.predict
pre = clf.predict(x_test)
```

```python
# step4 驗證模型(預測與ytest來比較正確率)
# get model accuracy
from sklearn import metrics
metrics.accuracy_score(pre, y_test)
```

```python
from sklearn.tree import export_graphviz
import graphviz
g = export_graphviz(clf, 
                    feature_names=iris["feature_names"],
                    class_names=iris["target_names"],
                    filled=True)
graphviz.Source(g)
```
